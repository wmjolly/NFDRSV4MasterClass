```python
#import libraries
    #C implementation of ET is faster if available
try:
	from xml.etree import cElementTree as ET
except ImportError:
	from xml.etree import ElementTree as ET
import pandas as pd
import urllib.request as urllib
#enable inline plotting
%pylab inline
```

```python
#define start and end dates for nfdrs data and weather data
    #does not need to be user input, use datetime and timedelta to automate
nfdrs_st = '01-JAN-2022'
nfdrs_end = '31-DEC-2022'
wx_st = '01-NOV-2022'
wx_end = '31-DEC-2022'
```

```python
#create a Station class to retrieve data from WIMS
    #Station has the following methods; .nfdrs(), .wx(), and .pfcst()
    #dates are converted to datetime objects and set as the index
class Station(object):

	def __init__(self, stn_id, priority):
		self.stn_id = stn_id
		self.priority = priority
		
	def wxml_to_df(self, url):
		xml_data = urllib.urlopen(url)
		root = ET.XML(xml_data.read())
		all_records = []
		for i, elem in enumerate(root):
			record = {}
			for child in elem:
				record[child.tag] = child.text
			all_records.append(record)
		df = pd.DataFrame(all_records)
		return df

	def nfdrs(self):
		url = ('https://famprod.nwcg.gov/wims/xsql/nfdrs.xsql?stn='
			+ self.stn_id + '&sig=&type=&start='
			+ nfdrs_st + '&end=' + nfdrs_end + '&time=&priority='
			+ self.priority + '&fmodel=&sort=&ndays=&user=FS10445')
		df = self.wxml_to_df(url)
		df.loc[:, 'nfdr_dt'] = pd.to_datetime(df['nfdr_dt'], format='%m/%d/%Y')
		df.set_index('nfdr_dt', inplace=True)
		return df

	def wx(self):
		url = ('https://famprod.nwcg.gov/wims/xsql/obs.xsql?stn='
			+ self.stn_id + '&sig=&type=&start=' + wx_st
			+ '&end=' + wx_end + '&time=&sort=&ndays=&user=FS10445')
		df = self.wxml_to_df(url)
		df.loc[:, 'obs_dttm'] = df.obs_dt.astype(str).str.cat(df.obs_tm.astype(str), sep=' ')
		df.loc[:, 'obs_dttm'] = pd.to_datetime(df['obs_dttm'], format='%m/%d/%Y %H')
		df.loc[:, 'obs_dt'] = pd.to_datetime(df['obs_dt'], format='%m/%d/%Y')
		df.set_index('obs_dt', inplace=True)
		return df

	def pfcst(self):
		url = ('https://famprod.nwcg.gov/wims/xsql/pfcst.xsql?stn='
			+ self.stn_id + '&sig=&type=&start=' + wx_st
			+ '&end=' + wx_end + '&time=&sort=&ndays=&user=FS10445')
		df = self.wxml_to_df(url)
		df.loc[:, 'fcst_dt'] = pd.to_datetime(df['fcst_dt'], format='%m/%d/%Y')
		df.set_index('fcst_dt', inplace=True)
		return df
```

```python
#create a Sig class to retrieve data from WIMS using Station class
    #for demo simplicity only the .nfdrs() method was added to Sig 
        #but in the same manner as the Station class .wx() and .pfcst() methods can be added here as well
    #text objects are converted to floats for defined columns
class Sig(object):

	def __init__(self, stn_list):
		self.stn_list = stn_list
		self.ec_breaks = []
		self.bi_breaks = []
    #the next 2 methods are not necessary, will explain use later
	def add_ec_breaks(self, breaks):
		self.ec_breaks.clear()    
		[self.ec_breaks.append(x) for x in breaks]

	def add_bi_breaks(self, breaks):
		self.bi_breaks.clear()
		[self.bi_breaks.append(x) for x in breaks]

	def nfdrs(self):
		l = []
		for station in self.stn_list:
			s = station[0]
			fm = station[1]
			stn = Station(s, fm)
			l.append(stn.nfdrs())
		df = pd.concat(l, axis=0)
        #dtypes need to be converted from objects for Sig mean
		for col in ['one_hr', 'ten_hr', 'hu_hr', 'th_hr', 'xh_hr', 'ic', 'kbdi', 'sc', 'ec', 'bi',
			'hrb', 'wdy', 'herb_gsi', 'woody_gsi']:
			df[col] = df[col].astype('float')
        #return 2 DataFrame Nelson and Forecast
        #could also return the station data here for the Sig
            #i.e. df_nelson = df.loc[df.nfdr_type == 'N'] 
		df_obs = pd.pivot_table(df.loc[df.nfdr_type == 'N'], index='nfdr_dt', aggfunc='mean')
		df_for = pd.pivot_table(df.loc[df.nfdr_type == 'F'], index='nfdr_dt', aggfunc='mean')
		return [df_obs, df_for]
```

```python
#create a Sig class to read data from daily listing text file
class Historic(object):

    def __init__(self, daily_listing):
        self.daily_listing = daily_listing

    def data(self):
        ffp = self.daily_listing
        df = pd.read_csv(ffp, header=0, engine='python', skipfooter=1)
        df = df.iloc[:, :-1]
        df = df.iloc[1:]
        df = df.rename(columns=lambda x: x.strip())
        df['DATE'] = pd.to_datetime(df['DATE'])
        return df
```

```python
#some of us are using percentiles for communication of indices (different from using for setting breakpoints)
    #this is one way to calculate using the Historic.data()
#this method returns a dictionary as {value: percentile value} pairs 
    #useful for adding a 2-nd y-axis to charts
def perc_yax(historic_data, indice):
	s = historic_data[indice].quantile(np.linspace(0.0,1.0,101)) #maybe this should start at 0.1?
	df = pd.DataFrame(s.reset_index())
	df = df.rename({'index':'percentile', indice: 'value'}, axis=1)
	df['percentile'] = df['percentile'] * 100
	df.value = df.value.astype(int)
	df.drop_duplicates(subset='value', keep='last', inplace=True)
	df.set_index(df.value, inplace=True, drop=True)
	df2 = pd.DataFrame({'value': range(0, df.value.max() + 1), 'percentile': ''})
	df2.set_index(df2.value, inplace=True, drop=True)
	df2.percentile = df.percentile
	df2.percentile = df2.percentile.ffill()
	df2.percentile = df2.percentile.bfill()
	df2.reset_index(drop=True, inplace=True)
	d = dict(zip(df2['value'], df2['percentile'].astype(int).astype(str)))
	return d

#this method takes a value and returns the percentile, uses the above function as var
    #allows for interpolation between dict k-v pairs
from scipy.interpolate import interp1d
def interp(perc_yax):
	x = list(perc_yax.keys())
	y = list(perc_yax.values())
	i = interp1d(x, y, fill_value='extrapolate')
	return i
```
